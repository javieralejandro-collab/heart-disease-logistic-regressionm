```python
#  Prueba Chi-cuadrado
from scipy.stats import chi2_contingency

# Variables categóricas para analizar
variables_categoricas = ['Sex', 'Chest pain type', 'FBS over 120', 'EKG results',
                         'Exercise angina', 'Slope of ST', 'Number of vessels fluro', 'Thallium']

print("Prueba Chi-cuadrado vs Enfermedad Cardíaca:\n")

for var in variables_categoricas:
    # Crear tabla de contingencia (como en tu código)
    contingency = pd.crosstab(df[var], df['Heart Disease'])
    chi2, p, dof, expected = chi2_contingency(contingency)

    print(f"🔹 {var}:")
    print(f"   Chi2 = {chi2:.4f}")
    print(f"   p-valor = {p:.2e}")

    if p < 0.05:
        print(f"   SIGNIFICATIVA (p < 0.05)")
    else:
        print(f"   NO significativa")
    print()
```

```python
#  Funciones de Regresión Logística

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def costo(y, p):
    return -np.mean(y * np.log(p) + (1 - y) * np.log(1 - p))
```

```python
def CoefRegLogistica(X, y, learning_rate=0.01, iteraciones=50000):
    m, n = X.shape
    X = np.c_[np.ones((m, 1)), X]  # Agregar columna de 1's para el intercepto
    beta = np.zeros(n + 1)

    for i in range(iteraciones):
        z = np.dot(X, beta)
        p = sigmoid(z)
        gradiente = np.dot(X.T, (p - y)) / m
        beta -= learning_rate * gradiente

        if i % 10000 == 0:
            cost = costo(y, p)
            print(f'Iteracion {i}, Costo: {cost}')

    return beta
```

```python
# Evaluar
def predecir(X, beta, umbral=0.5):
    m = X.shape[0]
    X_con_1 = np.c_[np.ones((m, 1)), X]
    probabilidad = sigmoid(np.dot(X_con_1, beta))
    return (probabilidad >= umbral).astype(int)

y_pred = predecir(X_test_scaled, beta)

# Matriz de confusión
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print("Matriz de Confusión:")
print(cm)
print(f"\nPresición: {(y_pred == y_test).mean():.4f}")
```
